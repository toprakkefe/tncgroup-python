```python
def gorev_oku():
  gorevlistesi = []
  try:
    with open("veritabani.txt","r",encoding="utf-8") as dosya:
      for veri in dosya:
        gorevlistesi.append(veri.strip())

    return gorevlistesi

  except FileNotFoundError:
    with open("veritabani.txt","w",encoding="utf-8") as dosya:
      pass
    print("\nDosya Başarıyla Oluşturuldu.")
    return []


def dosya_kayit(gorevlistesi):
  with open("veritabani.txt", "w", encoding="utf-8") as dosya:
        for gorev in gorevlistesi:
            dosya.write(gorev + "\n")

gorevler = gorev_oku()

def gorev_listele(gorevlistesi):
  print("\n--- GÖREV LİSTESİ ---")
  if not gorevler:
        print("\nListe henüz boş.")
  else:
      for no, gorev in enumerate(gorevler, start=1):
          print(f"{no}. {gorev}")

def gorev_ekle(gorevlistesi):

  yenigorev = input("\nYeni görev (Çıkmak için 'q'): ").strip()
  if yenigorev.lower() == 'q':
    return

  elif not yenigorev:
    print("\nHata! Boş görev giremezsiniz.")
    return

  elif yenigorev:
      gorevler.append(yenigorev)
      dosya_kayit(gorevler)
      print(f"\nGörev başarıyla {len(gorevler)}. numaraya eklendi!")


def gorev_sil(gorevlistesi):

    if not gorevler:
      print("\nListe henüz boş.")
      return

    try:
      secim = int(input(f"\nSilmek istediğiniz görev numarasını girin(1-{len(gorevlistesi)}): "))


      if 1 <= secim <= len(gorevlistesi):
          silinen = gorevler.pop(secim - 1)

          dosya_kayit(gorevlistesi)
          print(f"\n{silinen}' adlı görev başarıyla silindi.")

      else:
          print(f"\nHata! Geçersiz bir numara girdiniz. Yalnızca 1-{len(gorevlistesi)} arası seçim yapabilirsiniz.")

    except ValueError:
      print("\nHata! Lütfen sadece sayı giriniz.")


def gorev_duzenle(gorevlistesi):

  if not gorevler:
      print("\nListe henüz boş.")
      return

  try:
    secim = int(input(f"\nDüzenlemek istediğiniz görev numarasını girin(1-{len(gorevlistesi)}): "))


    if 1 <= secim <= len(gorevlistesi):
        yenigorev = input("\nYeni görev metnini girin: ").strip()

        if yenigorev:
          eskigorev = gorevlistesi[secim - 1]
          gorevlistesi[secim - 1] = yenigorev

          dosya_kayit(gorevlistesi)
          print(f"\n{eskigorev}' görevi '{yenigorev}' olarak güncellendi!")

    else:
        print(f"\nHata! Geçersiz bir numara girdiniz. Yalnızca 1-{len(gorevlistesi)} arası seçim yapabilirsiniz.")

  except ValueError:
    print("\nHata! Lütfen sadece sayı giriniz.")


def gorev_tamamla(gorevlistesi):

    if not gorevler:
      print("\nListe henüz boş.")
      return

    try:
        secim = int(input("\nTamamlanan görev numarasını girin: "))
        if 1 <= secim <= len(gorevlistesi):
            indeks = secim - 1

            if not gorevlistesi[indeks].startswith("[TAMAMLANDI]"):
              gorevlistesi[indeks] = "[TAMAMLANDI] " + gorevlistesi[indeks]
              dosya_kayit(gorevlistesi)
              print("\nGörev başarıyla tamamlandı olarak işaretlendi!")
            else:
                print("\nBu görev zaten tamamlanmış durumda.")
        else:
            print(f"\nHata! Geçersiz bir numara girdiniz. Yalnızca 1-{len(gorevlistesi)} arası seçim yapabilirsiniz.")
    except ValueError:
        print("\nHata! Lütfen sadece sayı giriniz.")

def temizlik(gorevlistesi):
    yeni_liste = [g for g in gorevlistesi if not g.startswith("[TAMAMLANDI]")]

    silinen_sayisi = len(gorevlistesi) - len(yeni_liste)

    if silinen_sayisi > 0:
        gorevlistesi.clear()
        gorevlistesi.extend(yeni_liste)


        dosya_kayit(gorevlistesi)
        print(f"\nTemizlik Tamamlandı! {silinen_sayisi} adet görev listeden kaldırıldı.")
    else:
        print("\nTemizlenecek tamamlanmış görev bulunamadı.")


def ana_menu():

      while True:
          print("\n=== ANA MENÜ ===")
          print("1. Görevleri Listele")
          print("2. Yeni Görev Ekle")
          print("3. Görev Düzenle")
          print("4. Görev Sil")
          print("5. Tamamlanmış Görev İşaretle/Temizle")
          print("6. Çıkış")

          secim = input("\nSeçiminiz: ")

          if secim == "1":
              gorev_listele(gorevler)
          elif secim == "2":
            gorev_ekle(gorevler)
          elif secim == "3":
            gorev_duzenle(gorevler)
          elif secim == "4":
            gorev_sil(gorevler)
          elif secim == "5":
            islem = input("\nHangi İşlemi Yapmak İstiyorsunuz?\n\n1. Görev İşaretle\n\n2.Görev Temizle\n\n")
            if islem == "1":
              gorev_tamamla(gorevler)
            elif islem == "2":
              temizlik(gorevler)
            else:
              print("\nHata! Yalnızca 1-2 seçeneklerini giriniz.")

          elif secim == "6":
              print("\nUygulama kapanıyor, hoşça kalın!")
              break
          else:
              print("\nHata! Yalnızca 1-5 arasında bir rakam giriniz.")


ana_menu()








```
